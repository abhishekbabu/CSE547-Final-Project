```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
val = []
test = []
with open('./Appliances_default/log.txt') as file:
    for line in file:
        line = line.replace('(', '')
        line = line.replace(')', '')
        line = line.replace(',', '')
        line = line.split()
        ndcg_val = float(line[0])
        ndcg_test = float(line[2])
        val.append(ndcg_val)
        test.append(ndcg_test)
```

```python
epoch = np.arange(len(val)) * 20
```

```python
sns.set_theme()
plt.figure(figsize=(10, 6))
plt.plot(epoch, val, '-o', label='Validation')
plt.plot(epoch, test, '-o', label='Te')
```
